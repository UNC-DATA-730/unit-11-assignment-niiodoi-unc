Let's build a random forest model to predict `MedianHouseValue`.

1. Create your model, set the engine to `ranger` with `importance = 'impurity'` and the mode as `regression`.
2. Fit the model using `MedianHouseValue` as the response variable and all other columns as predictors.

````r
# Create scatter plot
ggplot(housing, aes(x = MedInc, y = MedianHouseValue)) +
  geom_point()
````

````r
# Define random forest model

rf_mod = rand_forest() |>
    set_engine("ranger", importance= "impurity") |>
    set_mode("regression")

model_fit <- rf_mod |>
    fit(MedianHouseValue ~ ., data = housing)

model_fit
````

````r
rec = recipe(MedianHouseValue ~ ., data = housing) |>
    step_normalize(all_numeric_predictors()) |>
    step_impute_mean(all_numeric_predictors()) |>
    prep()
rec |> juice() |> head()
````

**Answer:**

The OOB error for this model is 0.231802. The model R-squared value is 0.8259 which indicates that the model explains approximately 82.59% of the variability in the data.
